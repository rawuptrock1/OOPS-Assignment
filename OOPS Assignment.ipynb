4  : - What is the difference between abstraction and encapsulation?
---------------------------------------------------------
Both are important OOP concepts, but they serve different roles:
Abstraction refers to a concept of hiding complex implementation details and showing only the necessary parts to the user.
Encapsulation is about bundling data and methods that operate on that data within one unit, like a class, and restricting access to some of the object's components.

 Eg: =
---------

```python

#2. Write a program to create an abstract class Shape with a method area(). Derive classes Circle and Rectangle from it and implement the area() method in both.
class Shape:        # base "abstract" style class
    def area(self):       # method intended to be overridden
        raise NotImplementedError("Subclasses must implement this method")  # enforce override

class Circle(Shape):        # Circle derives from Shape
    def __init__(self, radius):      # constructor takes radius parameter
        self.radius = radius    # store radius in instance attribute

    def area(self):        # override area for circle
        return 3.14 * self.radius * self.radius  # return area using pi ~ 3.14

class Rectangle(Shape):    # Rectangle derives from Shape
    def __init__(self, width, height):   # constructor takes width and height
        self.width = width                     # store width
        self.height = height       # store height

    def area(self):              # override area for rectangle
        return self.width * self.height  # return width * height
rad=int(input("Enter radius of the circle: "))
c = Circle(rad) 
length=int(input("Enter length of the rectangle: "))
width=int(input("Enter width of the rectangle: "))
r = Rectangle(length, width)      # Rectangle with user-defined dimensions
print("Circle Area:", c.area())       # prints computed circle area
print("Rectangle Area:", r.area())   # prints computed rectangle area
```

```python

#5. Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes 
# balance and methods to deposit, withdraw, and check balance.
class BankAccount:                              # BankAccount class to show encapsulation
    def __init__(self, initial_balance=0):      # constructor with optional initial balance
        self.__balance = initial_balance       # private attribute, name-mangled to _BankAccount__balance

    def deposit(self, amount):                  # method to deposit money
        if amount > 0:                          # validate positive amount
            self.__balance += amount           # modify private balance
            print(f"Deposited: {amount}")      # print confirmation
        else:                                   # invalid deposit
            print("Deposit must be positive.") # print error message

    def withdraw(self, amount):                 # method to withdraw money
        if 0 < amount <= self.__balance:        # check sufficient funds and positive amount
            self.__balance -= amount           # reduce balance
            print(f"Withdrew: {amount}")       # confirmation
        else:                                   # invalid withdrawal or insufficient funds
            print("Insufficient balance or invalid amount.")  # print error

    def check_balance(self):                    # accessor method to inspect balance
        print("Current Balance:", self.__balance)  # prints the private balance
        return self.__balance                   # return balance value
opening_balance=1000
acct = BankAccount(opening_balance)            # create account with initial balance
acct.deposit(int(input("Enter deposit amount: ")))  # deposit user-defined amount
withdrawal_amount = int(input("Enter withdrawal amount: "))
if withdrawal_amount > acct.check_balance():
    print("Insufficient balance for withdrawal.")
else:
    acct.withdraw(withdrawal_amount)
acct.check_balance()                            # show resulting balance

```

```python

#7. Create a class MathOperations with a class method add_numbers() to add two numbers and a static method subtract_numbers() to subtract two numbers.
class MathOperations:  # utility class for simple math ops
    @classmethod                   # decorator marking a class method
    def add_numbers(cls, a, b):     # class method receives class as first arg
        return a + b        # return sum

    @staticmethod          # decorator for static method
    def subtract_numbers(a, b):        # static method does not receive cls or self
        return a - b        # return difference
num1 = int(input(" enter a number (num1) :"))
num2 = int(input(" enter another number (num2) :"))
print("Addition:", MathOperations.add_numbers(num1, num2))    # call class method
print("Subtraction:", MathOperations.subtract_numbers(num1, num2))  # call static method
```

```python

#10. Demonstrate operator overloading by creating a class Vector and overriding the add method to add two vectors.
class Vector:                                  # simple 2D vector class
    def __init__(self, x, y):                  # constructor
        self.x = x
        self.y = y
    def __add__(self, other):                  # overload + operator
        if isinstance(other, Vector):
            return Vector(self.x + other.x, self.y + other.y)
        return NotImplemented
    def __str__(self):                         # readable display format
        return f"Vector({self.x}, {self.y})"

x1 = int(input("Enter x1 for first vector: "))
y1 = int(input("Enter y1 for first vector: "))

x2 = int(input("Enter x2 for second vector: "))
y2 = int(input("Enter y2 for second vector: "))

v1 = Vector(x1, y1)    # Create first vector 
v2 = Vector(x2, y2)    # Create second vector

v3 = v1 + v2           # uses overloaded __add__
print("v1 + v2 =", v3) # Output the result
```

```python

#12. Implement a class Student with attributes name and grades. Create a method average_grade() to compute the average of the grades.
class StudentGrades:              # class to hold student name and grades
    def __init__(self, name):    # constructor with name
        self.name = name         # store name
        self.grades = []      # initialize empty list for grades

    def add_grade(self, grade):  # method to add a grade
        self.grades.append(grade)   # append grade to the list

    def average_grade(self):    # compute average grade
        return sum(self.grades) / len(self.grades) if self.grades else 0  # avoid division by zero

name = input("Enter Student Name: ")
s = StudentGrades(name)  # create student
while True:
    grade_input = input("Enter grade (or type 'done' to stop): ")
    if grade_input.lower() == 'done':
        break
    if grade_input.isdigit():  # ensure numeric input
        s.add_grade(int(grade_input))
    else:
        print("Please enter only numbers for grade!") 
print("Average Grade for", s.name, "is", s.average_grade())  # print average
```

```python

#13. Create a class Rectangle with methods set_dimensions() to set the dimensions and area() to calculate the area.
class RectangleClass:       # Rectangle class example
    def __init__(self):     # default constructor
        self.width = 0      # defalt width 0
        self.height = 0     # default height 0

    def set_dimensions(self, width, height):    # setter method to set dimensions
        self.width = width      # assign width
        self.height = height      # assign height

    def area(self):   # method to compute area
        return self.width * self.height        # return width * height

rect = RectangleClass()      # create RectangleClass instance
w = float(input("Enter the width of rectangle: "))
h = float(input("Enter the height of rectangle: "))
rect.set_dimensions(w, h)       # set width and height

print("Rectangle area:", rect.area())  # prints 50
```

```python

#14. Create a class Employee with a method calculate_salary() that computes the salary based on hours worked and hourly rate. Create a derived class Manager that adds a bonus to the salary.
class EmployeeClass:  # base Employee class
    def __init__(self, name, hours_worked, hourly_rate):  # constructor with payroll details
        self.name = name
        self.hours_worked = hours_worked
        self.hourly_rate = hourly_rate

    def calculate_salary(self):  # base salary calculation
        return self.hours_worked * self.hourly_rate  # multiply hours by rate

class ManagerClass(EmployeeClass): # Manager extends Employee
    def __init__(self, name, hours_worked, hourly_rate, bonus):  # add bonus param
        super().__init__(name, hours_worked, hourly_rate)   # initialize base part
        self.bonus = bonus     # store manager bonus

    def calculate_salary(self):    # override to include bonus
        base_salary = super().calculate_salary() # compute base salary from Employee
        return base_salary + self.bonus  # add manager bonus
name = input("Enter Name: ")
hours = float(input("Enter working hours: "))
rate = float(input("Enter hourly rate: "))
bonus = float(input("Enter bonus: "))

mgr = ManagerClass(name, hours, rate, bonus)
print("Total Salary for", mgr.name, ":", mgr.calculate_salary())  # prints salary + bonus


```

```python

#15. Create a class Product with attributes name, price, and quantity. Implement a method total_price() that calculates the total price of the product.

class ProductClass:           # product representation
    def __init__(self, name, price, quantity):       # constructor with product details
        self.name = name
        self.price = price
        self.quantity = quantity

    def total_price(self):    # compute total price = price * quantity
        return self.price * self.quantity

name = input("Enter Product Name: ")
price = float(input("Enter Product Price: "))
quantity = int(input("Enter Product Quantity: "))

product = ProductClass(name, price, quantity)
print("Total price for", product.name, ":", product.total_price())  # prints total price                       

```

```python

#16. Create a class Animal with an abstract method sound(). Create two derived classes Cow and Sheep that implement the sound() method.
class AnimalSound:        # base class for sound demonstration
    def sound(self):          # method to be implemented by subclasses
        raise NotImplementedError("Subclasses must implement this method")  # enforcement

class Cow(AnimalSound):     # Cow implements AnimalSound
    def sound(self):             # override sound
        return "Mooooo"         # return cow sound

class Sheep(AnimalSound):       # Sheep implements AnimalSound
    def sound(self):           # override sound
        return "Baa...baa"     # return sheep sound


print("Cow:", Cow().sound())      # prints "Moo"
print("Sheep:", Sheep().sound())    # prints "Baa"
```

```python

#17. Create a class Book with attributes title, author, and year_published. Add a method get_book_info() that returns a formatted string with the book's details.

class BookClass:                                     # Book representation class
        def __init__(self, title, author, year_published):  # constructor with book details
            self.title = title # store title
            self.author = author     # store author
            self.year_published = year_published   # store year published
    
        def get_book_info(self):    # return formatted book info
            return f"'{self.title}' by {self.author}, published in {self.year_published}."  # formatted string
def get_valid_year():
    while True:
        year_input = input("Enter Publication Year): ")
        if year_input.isdigit() and len(year_input) == 4:
            return int(year_input)
        else:
            print("Invalid input! Please enter a valid  year (e.g., 1999, 2024).")

aut = input("Enter Book Author: ")
title = input("Enter Book Title: ")
year = get_valid_year()


book = BookClass(title, aut, year) # instantiate BookClass
print(book.get_book_info())  # prints formatted book info
```

```python
#18. Create a class House with attributes address and price. Create a derived class Mansion that adds an attribute number_of_rooms.

class HouseClass:                                    # base House class
        def __init__(self, address, price):             # constructor with address and price
            self.address = address                       # store address
            self.price = price
                                     
    
class MansionClass(HouseClass):                      # Mansion inherits from House
        def __init__(self, address, price, number_of_rooms):  # constructor adds number_of_rooms
            super().__init__(address, price)            # initialize base class
            self.number_of_rooms = number_of_rooms      # store number of rooms
    
address = input("Enter Mansion Address: ")
while True:
    price_input = input("Enter Mansion Price (numbers only): ")
    if price_input.isdigit():
        price = int(price_input)
        break
    else:
        print("Invalid! Please enter only numbers for price.")
while True:
    rooms_input = input("Enter Number of Rooms: ")
    if rooms_input.isdigit():
        rooms = int(rooms_input)
        break
    else:
        print(" Invalid! Only whole numbers allowed for rooms.")   

mansion = MansionClass(address, price, rooms)  # create Mansion instance
print("Address:", mansion.address, "| Price:", mansion.price, "| Rooms:", mansion.number_of_rooms)  # print details
```
